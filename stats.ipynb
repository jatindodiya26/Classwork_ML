```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt


np.random.seed(42)


#simukate dataset
data={
    "math_score":np.random.normal(70,10,100),
    "Science_score":np.random.normal(72,12,100),
    "Hours_Studied":np.random.normal(5,2,100)
}

df=pd.DataFrame(data)
print(df.head(),"/n")






mean_math=df["math_score"].mean()
median_math=df["math_score"].median()
mode_math=stats.mode(df["math_score"],keepdims=True)[0][0]

print("mean_path:",mean_math)
print("median_math:",median_math)
print("mode (math):",mode_math, "\n")




#data distribution
plt.figure(figsize=(10,5))
plt.hist(df["math_score"],bins=20,color='skyblue',edgecolor="black")
plt.title("math score distribution")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show()

plt.figure(figsize=(10,5))
plt.boxplot(df["math_score"])
plt.title("math score distribution")
plt.show()
```

# **ML**

Supervised machine learning, or supervised learning, is a type of machine learning (ML) used in artificial intelligence (AI) applications to train algorithms using labeled datasets.


Classification: A supervised learning technique used to predict discrete, categorical labels or classes (e.g., "spam" vs. "not spam", "true" vs. "false").


Regression: A supervised learning technique used to predict continuous, numerical values (e.g., house prices, temperature).


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
hours_studied = np.array([2, 4, 6]).reshape(-1, 1)
marks = np.array([40, 50, 60])

# Create and train model
model = LinearRegression()
model.fit(hours_studied, marks)

# Predictions (best-fit line)
predicted_marks = model.predict(hours_studied)

# Plot data points
plt.scatter(hours_studied, marks, color='blue', label='Actual Data')

# Plot regression line
plt.plot(hours_studied, predicted_marks, color='red', label='Best Fit Line')

# Labels and title
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.title('Linear Regression - Best Fit Line')
plt.legend()

# Show plot
plt.show()

# Print equation
slope = model.coef_[0]
intercept = model.intercept_
print(f"Equation: Marks = {slope:.2f} * Hours + {intercept:.2f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
experience=np.random.normal(1,11,50)
salary=experience*8+np.random.normal(0,3,50)
```

```python
df=pd.DataFrame({"experienece ": experience, "salary":salary})
```

```python
print(df.head())
```

```python
x=df[['experienece ']]
y=df['salary']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
print("Intercept", model.intercept_)
print("Slope", model.coef_[0])
```

```python
y_pred=model.predict(x_test)
```

```python
MAE=mean_absolute_error(y_test,y_pred)
MSE=mean_squared_error(y_test,y_pred)
RMSE=np.sqrt(MSE)
r2_score=r2_score(y_test,y_pred)
```

```python
print(MAE)
print(MSE)
print(RMSE)
print(r2_score)

```

```python
plt.scatter(x,y,color='blue',label="actual data")
plt.plot(x,model.predict(x),color='red', label= 'regression line')
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.title("Salary vs Experience")
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,color='red')
plt.legend()
plt.xlabel("Experience")
plt.ylabel("Salary")

plt.show()
```
