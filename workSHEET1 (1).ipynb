1. OUTLIER DETECTION (IQR METHOD)
a. Define Q1, Q3, and IQR in your own words.
b. Given data: [5, 7, 10, 12, 13, 15, 200], calculate:
- Q1, Q3, IQR
- Lower and Upper Bound
- Identify outliers

Q1- the value below which 25% of data falls
Q3- the value below which 75% of data falls
IQR - the spread of 50% of data


```python
import pandas as pd
```

```python
data= [5, 7, 10, 12, 13, 15, 200]
```

```python
df=pd.DataFrame(data,columns=['data'])
Q1=df['data'].quantile(0.25)
Q3=df['data'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["data"] < lower_bound) | (df["data"] > upper_bound)]
print("Outliers:\n", outliers)

```

2. FEATURE SCALING
a. What is the difference between Normalization and Standardization?
b. Apply Min-Max Scaling to: [10, 20, 30]

Normalization scales data between range 0 to 1
and standardization makes data 0 centered

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform([[10],[20],[30]])
print(scaled)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform([[5],[15],[25]])
scaled
```
